# Create Train/Test Dataset

Copy 100 images and create the train dataset, and 20 images for the test dataset


```python

import os
import random
import shutil



dataDir = 'data/Images'
trainDir = 'data/train'
testDir= 'data/test'

def checkDirectoryExists(newDir):
    #print(newDir)
    if not os.path.exists(newDir):
        os.makedirs(newDir)

def copyFile(fileName,sourceDirectory,destinationDirectory):
    newFileName=os.path.join(destinationDirectory,fileName)
    #print("New File :",newFileName)
    copyFileName=os.path.join(sourceDirectory,fileName)
    if not os.path.exists(newFileName):
        shutil.copy(copyFileName, destinationDirectory)
        return True
    return False
                    
random.seed(42)        
for _, dirs, _ in os.walk(dataDir):
    for dir in dirs:
        #print(dir)
        copyTrainDir = os.path.join(trainDir, dir)
        checkDirectoryExists(copyTrainDir)
        
        copyTestDir = os.path.join(testDir, dir)
        checkDirectoryExists(copyTestDir)
        copydir=os.path.join(dataDir, dir)
        
        #print("Train Directory :",copyTrainDir)
        #print("Test Directory :",copyTestDir)
            
        for _, _, files in os.walk(copydir):
            # Randomize file list (may not be necessary)
            print("Copy Directory :",copydir)
            random.shuffle(files)
            #copy 100 files to Train
            for f in files[:100] :
                copyFile(f,copydir,copyTrainDir)
            #copy 20 files to Test 
            for f in files[100:120] :
                copyFile(f,copydir,copyTestDir)   

            
            


        
```
